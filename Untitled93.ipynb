```python
import PyPDF2
```

```python
def load_pdf(file_path):
    with open(file_path, 'rb') as f:
        reader = PyPDF2.PdfReader(f)
        text = ''
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text()
    return text
```

```python
document_text = load_pdf('/content/Copy of Bento Design Apple Keynote Inspired Background Slides_compressed.pdf')
```

```python
from transformers import AutoTokenizer, AutoModel
```

```python
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
model = AutoModel.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
```

```python
def get_embeddings(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        embeddings = model(**inputs).last_hidden_state.mean(dim=1).squeeze().cpu().numpy()
    return embeddings
```

```python
document_chunks = [document_text[i:i+500] for i in range(0, len(document_text), 500)]
```

```python
def upload_document_chunks(chunks):
    for idx, chunk in enumerate(chunks):
        embedding = get_embeddings(chunk)
        client.data_object.create(
            {
                "content": chunk
            },
            "Document",
            vector=embedding.tolist()
        )

upload_document_chunks(document_chunks)
```

```python
def query_answer(query):
    query_embedding = get_embeddings(query)
    result = client.query.get("Document", ["content"])\
                .with_near_vector({"vector": query_embedding.tolist()})\
                .with_limit(3)\
                .do()

    return result
```

```python
response = query_answer("What is the business model?")
print(response)
```

```python
def generate_response(context, query):
    response = cohere_client.generate(
        model='command',
        prompt=f"Context: {context}\n\nQuestion: {query}?\nAnswer:",
        max_tokens=100
    )
    return response.generations[0].text.strip()
```

```python
context = ' '.join([doc['content'] for doc in response['data']['Get']['Document']])
```
